```python
from google.colab import files
from pathlib import Path

uploaded = files.upload()
model_path = next((name for name in uploaded if name.lower().endswith(('.pt', '.onnx'))), None)
image_path = next((name for name in uploaded if name.lower().endswith(('.jpg', '.jpeg', '.png', '.webp'))), None)

if model_path is None:
    raise ValueError('Upload a model file named best.pt or best.onnx.')
if image_path is None:
    raise ValueError('Upload a test image such as photo.jpg.')

print('Model:', model_path)
print('Image:', image_path)
```

```python
from pathlib import Path

output_dir = Path('/content/runs/colab_test')
output_dir.mkdir(parents=True, exist_ok=True)

!python detect.py --weights "$model_path" --source "$image_path" --img 640 --conf 0.5 --project /content/runs/colab_test --name result --exist-ok
```
